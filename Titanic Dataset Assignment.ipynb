**TITANIC Dataset Visualization**

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic Dataset
data = pd.read_csv('titanic.csv')
```

```python
#Display the first few rows of the dataset
data.head()
```

```python
### Question 1: What is the distribution of passenger ages on the Titanic?
## Plottting age distribution of passengers
sns.histplot(data['Age'], bins=20, kde=True, color='Purple')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Inference: Most passengers are between the ages of 20-40 years
```

```python
### Question 2: How does the survival rate vary based on passenger class (Pclass)?
## Plotting Survival rate based on Pclass
sns.barplot(x='Pclass', y='Survived', data=data, color='Blue')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Inference: Survival rate of Passenger class 1 is the highest.
```

```python
### Question 3: Is there a relationship between the fare paid and the likelihood of survival?
## Plotting Survival based on fare paid
sns.boxplot(x='Survived', y='Fare', data=data, palette ='pastel')
plt.title('Survival vs. Fare')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

```python
#Inference: Higher fare paid has higher chances of survival
```

```python
### Question 4: How does the survival rate differ between male and female passengers?
## Plotting Survival rate by Gender
sns.barplot(x='Sex', y='Survived', data=data, palette='pastel')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Inference: Females have a higher survival rate.
```

```python
### Question 5: What impact does the number of siblings/spouses (SibSp) or parents/children(Parch) aboard have on survival?
## Plotting Survival based on siblings/spouses
sns.countplot(x='SibSp', hue='Survived', data=data, palette='Spectral')
plt.title('Survival vs. Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Count')
plt.show()
```

```python
# Inference: Passengers with 1 siblings/spouses have higher survival chances
```

```python
## Plotting Survival based on parents/children
sns.countplot(x='Parch', hue='Survived', data=data, palette='RdBu')
plt.title('Survival vs. Number of Parents/Children Aboard')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Count')
```

```python
# Inference: Passengers with 1 and 3 Parents/Children had higher Survival chances.
```

```python
### Question 6: Are there any significant differences in survival rates among passengers embarked from different ports?
## Plotting Survival based on Embarked
sns.countplot(x='Embarked', hue='Survived', data=data, palette='PRGn')
plt.title('Survival vs. Port of Embarkation')
plt.xlabel('Port of Embarkation')
```

```python
# Inference: Passengers who embarked from Port 'C' have higher chances of survival.
```
