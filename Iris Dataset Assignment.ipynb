**IRIS DATASET ASSIGNMENT**

```python
#Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris Dataset
data = pd.read_csv('Iris.csv')
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
data.tail()
```

```python
###Question 1: What is the distribution of sepal lengths and widths among the Iris species?
##Plotting the distribution of Sepal Lengths and Sepal Widths
#Box Plot for Sepal Length
sns.boxplot(x='Species', y='SepalLengthCm', data=data, palette='coolwarm')
plt.title('Distribution of Sepal Lengths')
plt.show()
```

```python
#Inference: The Virginica Species have the Longest Sepal Length among all 3 species
```

```python
#Box plot for Sepal Width
sns.boxplot(x='Species', y='SepalWidthCm', data=data, palette='pastel')
plt.title('Distribution of Sepal Widths')
plt.show()
```

```python
#Inference: The Setosa Species have the Widest Sepal Width among all 3 species
```

```python
###Question 2: How does the petal length vary across different Iris species?
##Plotting the Petal Lengths
#Box Plot for Petal Length
sns.boxplot(x='Species', y='PetalLengthCm', data=data, palette='Spectral')
plt.title('Distribution of Petal Lengths')
plt.show()
```

```python
#Inference: The Virginica Species have the Longest Petal Length among all 3 species
```

```python
###Question 3: Is there a relationship between sepal length and petal length?
##Plotting Sepal and Petal lengths among Species
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=data, palette='coolwarm')
plt.title('Relationship between Sepal Length and Petal Length')
plt.show()
```

```python
#Inference: Setosa species has the smallest Sepal and Petal Lengths and Virginica species has the largest Sepal and Petal lengths. The relationship between sepal and petal lengths is Linear

```

```python
###Question 4: How does the petal width vary across different Iris species?
##Plotting the Petal Widths
#Box Plot for Petal Width
sns.boxplot(x='Species', y='PetalWidthCm', data=data, palette='pastel')
plt.title('Distribution of Petal Widths')
plt.show()
```

```python
#Inference: The Virginica species has the Widest Petal width among all 3 species.
```

```python
###Question 5: Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?
##Pairplot for pairwise relationship[s]
sns.pairplot(data, hue='Species', palette='Spectral')
plt.show()
```

```python
# Remove the 'Id' column from the dataset
data_without_id = data.drop(columns=['Id'])

# Create the pairplot without the 'ID' column
sns.pairplot(data_without_id, hue='Species', palette='Spectral')
plt.show()
```

```python
#Inference: There is a strong linear relationship between Petal Length and Petal width, Setosa species has the smallest petal dimensions and Virginica has the larget petal dimensions
```

```python
###Question 6: Can we distinguish between Iris species based on sepal and petal measurements?
##Plot Sepal and Petal dimensions by species
# Remove the 'Id', column from the dataset
data_without_id = data.drop(columns=['Id'])

# Create the pairplot without the 'ID' column
sns.pairplot(data_without_id, hue='Species', palette='Spectral')
plt.show()
```

```python
#Inference: The Petal measurements help in clearly distinguishing between differenmt species as we can observe clear separate clusters for each of the species and the least overlap between them.
```
